# KIVY

__A. PENGERTIAN KIVY__

__*Kivy*__ merupakan framework yang dibangun menggunakan library dari bahasa pemrograman Python yang bersifat Open Source. Tujuan dikembangkannya framework ini agar dapat membantu developer secara cepat dalam mengembangkan aplikasi yang memiliki tampilan antarmuka inovatif seperti aplikasi yang mendukung multitouch.

__B. CARA INSTALASI KIVY__

Sebelum menginstal framework kivy pada library python usahan terlebih dahulu sudah menginstal pip. Dan ini beberapa tahapan yang harus dilakukan untuk mengintal kivy melalui CMD :
1.	Ketik pada CMD python –m pip install –upgrade pip wheel setuptools.
2.	Kemudian selanjutnya ketik python –m pip install docutils pygments pypiwin32
Kivy.deps.sdl2 kivy.deps.glew pada proses ini nanti akan mendownload sebuah file berukuran kurang lebih 120 mb. Setelah download sudah selesai kemudian ketikkan pada kolom berikutnya Python –m pip install kivy.deps.gstreame.
3.	Selanjutnya yaitu proses instalasi kivy dengan cara ketikkan python –m pip install kivy kemudian tunggu sampai proses tersebut selesai dan muncul tulisan successfully installed kivy.

2.Pada pemrograman draw terdapat beberapa class diantaranya adalah :

    •	Class menggambar
        Pada class menggambar terdapat beberapa property pembangun yang nantinya akan disesuaikan dengan tampilan
        program yang akan ditampilkan,terdapat :
        1.Def on_touch_down
            Pada definisi  disini terdapat gambaran program yang nantinya akan melakukan gerakan kursor kearah bawah
            yang akan menjadi garis gambar pada layar 
        2.Def on_touch_move
            Pada definisi  disini terdapat gambaran program yang nantinya akan melakukan gerakan kursor kearah samping
            yang akan menjadi garis gambar pada layar 
        3.Def on_touch_up
            Pada definisi  disini terdapat gambaran program yang nantinya akan melakukan gerakan kursor kearah atas
            yang akan menjadi garis gambar pada layar 
    •	Class GambarkuApp menmpilkan beberapa button yang akan ditampilkan di layar sesuai dengan kebutuhan
        gambarnya,terdapat :
        1.Def build
            Pada definisi ini sendiri menjelaskan tentang button yang digunaka untuk menghapus gambar dan mengganti
            warna pada gambar dilayar secara acak 
        2.Def clear_canvas untuk menghapus gambar
            Dan pada akhir program terdapat GambarkuApp() yang digunakan untuk meamnggil keseluruhan program untuk
            ditampilkan pada layar mulai dari proses menggambar hingga terdapat tombol hapus dan random warna didalam
            layar.
